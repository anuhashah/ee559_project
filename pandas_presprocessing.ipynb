```python
import pandas as pd

train_set = "./laptop_data_train.csv"
test_set = "./laptop_data_test.csv"

train_data = pd.read_csv(test_set, header=0)
attributes = train_data.columns.tolist()

for attribute in attributes:
    if attribute != 'Number':
        print(f'Attribute: {attribute}\n {set(train_data[attribute])}\n\n')

      
        
new_columns = train_data['Cpu'].str.split(' ', expand=True)

# Fill blank substrings with NaN
new_columns = new_columns.replace('', pd.NA)

# Assign the new columns to the DataFrame
train_data[new_columns.columns] = new_columns

new_columns = train_data['ScreenResolution'].str.split(' ', expand=True)

# Fill blank substrings with NaN
new_columns = new_columns.replace('', pd.NA)

# Assign the new columns to the DataFrame
train_data[new_columns.columns] = new_columns

new_columns = train_data['Memory'].str.split(' ', expand=True)

# Fill blank substrings with NaN
new_columns = new_columns.replace('', pd.NA)

# Assign the new columns to the DataFrame
train_data[new_columns.columns] = new_columns

new_columns = train_data['Gpu'].str.split(' ', expand=True)

# Fill blank substrings with NaN
new_columns = new_columns.replace('', pd.NA)

# Assign the new columns to the DataFrame
train_data[new_columns.columns] = new_columns

# Drop the original column if needed
# df.drop(columns=['column_name'], inplace=True)

# Display the DataFrame with the new columns
# print(train_data)

styled_df = train_data.style

# Apply various formatting options
styled_df = styled_df.set_properties(**{'text-align': 'center'})  # Align text to center
styled_df = styled_df.set_table_styles([{'selector': 'th', 'props': [('font-size', '11pt')]}])  # Set font size for headers

# Display the styled DataFrame
styled_df
```
